```python
import pandas as pd
import numpy as np

np.random.seed(1)

data = {
    "Order_ID": range(1001, 1051),
    "Order_Date": pd.date_range(start="2024-01-01", periods=50, freq="7D"),
    "Product": np.random.choice(
        ["Laptop", "Mobile", "Tablet", "Headphones", "Smartwatch"], 50
    ),
    "Category": np.random.choice(
        ["Electronics", "Accessories"], 50
    ),
    "Quantity": np.random.randint(1, 6, 50),
    "Price": np.random.randint(500, 50000, 50)
}

df = pd.DataFrame(data)
df.head()

df["Total_Sales"] = df["Quantity"] * df["Price"]
df.head()

total_revenue = df["Total_Sales"].sum()
total_orders = df["Order_ID"].count()

print("Total Revenue:", total_revenue)
print("Total Orders:", total_orders)

product_sales = df.groupby("Product")["Total_Sales"].sum().sort_values(ascending=False)
product_sales

product_sales.idxmax()
product_sales.idxmin()

category_sales = df.groupby("Category")["Total_Sales"].sum()
category_sales

df["Month"] = df["Order_Date"].dt.month

monthly_sales = df.groupby("Month")["Total_Sales"].sum()
monthly_sales

print("""
Business Summary:
- Electronics category generates higher revenue.
- Laptop and Mobile are top-performing products.
- Monthly sales show steady demand throughout the year.
- Focus on high-revenue products can increase profitability.
""")

df.to_csv("week2_sales_analysis_output.csv", index=False)
print("CSV file exported successfully")
```
